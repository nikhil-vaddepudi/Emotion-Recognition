# data and preprocessing

We taken FER 2013 emotions daataset and we created a data list containing HOG Features and labels of each image. We stored 

final data list into the pickle file for easy computation.

```python
import os

import numpy as np

import cv2

import matplotlib.pyplot as plt

from skimage.feature import hog    #To get the HOG Features of an image.

import pickle                      
```

```python


dir='*t'

categories = [ 'angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise' ]

data=[]

for category in categories:
    
    path=os.path.join(dir,category)
    
    label=categories.index(category)
    
    for img in os.listdir(path):
        
        imgpath = os.path.join(path,img)
        
        emo_image=cv2.imread(imgpath,0)
        
        plt.imshow(emo_image)
        plt.show()
        
        break
        
    break

    
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
data=[]

for category in categories:
    
    path=os.path.join(dir,category)
    
    label=categories.index(category)
    
    for img in os.listdir(path):
        
        imgpath = os.path.join(path,img)
        
        emo_image=cv2.imread(imgpath,0)
        try:
            [hog_fv,hog_image]=hog(emo_image,orientations=9,pixels_per_cell=(8,8),cells_per_block=(2,2),visualize=True)
        
            data.append([hog_fv,label])
        except Exception as e:
            pass
        
        
```
