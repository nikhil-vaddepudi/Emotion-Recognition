```python
import numpy as np

import cv2

import matplotlib.pyplot as plt

from skimage.feature import hog

import pickle

import random
from sklearn import tree

from sklearn.model_selection import train_test_split

from sklearn.svm import SVC

from sklearn.metrics import accuracy_score

import pandas as pd

from sklearn import preprocessing
```

```python
pick=open('data1.py','rb')
data=pickle.load(pick)
pick.close()
```

```python
print(len(data))
```

```python
random.shuffle(data)

features=[]

labels=[]

for feature ,label in data:
    features.append(feature)
    labels.append(label)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, labels, test_size=0.20)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(xtrain, ytrain)
```

```python
clf.score(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

```python
y_hat = clf.predict(xtest)

accu= accuracy_score(ytest,y_hat)
print(accu)
```

```python
train_sco=[]
val_sc=[]
for i in range(1,10):
    clf = tree.DecisionTreeClassifier(max_depth=i)
    clf.fit(xtrain, ytrain)
    train_sco.append(clf.score(xtrain,ytrain))
    val_sc.append(clf.score(xtest,ytest))
   
```

```python
frame=pd.DataFrame({'max_depth':range(1,10),'train_accuracy':train_sco,'valid_sco':val_sc})
frame.head()
```

```python
plt.plot(frame['max_depth'],frame['train_accuracy'])
plt.plot(frame['max_depth'],frame['valid_sco'])
plt.show()
```

# max_depth=8

```python
train_sco=[]
val_sc=[]
for i in range(2,100):
    clf = tree.DecisionTreeClassifier(max_leaf_nodes=i)
    clf.fit(xtrain, ytrain)
    train_sco.append(clf.score(xtrain,ytrain))
    val_sc.append(clf.score(xtest,ytest))
   
```

```python
frame=pd.DataFrame({'max_leaf_nodes':range(2,100),'train_accuracy':train_sco,'valid_sco':val_sc})
frame.head()
```

```python
plt.plot(frame['max_leaf_nodes'],frame['train_accuracy'])
plt.plot(frame['max_leaf_nodes'],frame['valid_sco'])
plt.show()
```
