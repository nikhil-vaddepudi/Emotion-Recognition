```python
import os

import numpy as np

import cv2

import matplotlib.pyplot as plt

from skimage.feature import hog

import pickle

import random

from sklearn.model_selection import train_test_split

from sklearn.svm import SVC

from sklearn.metrics import accuracy_score

import pandas as pd

from sklearn import preprocessing
       
```

```python
# Retrieving data from a pickle file which is the final product of preprocessing

pick=open('data1.py','rb')
data=pickle.load(pick)
pick.close()
```

```python
print(len(data))
```

```python
random.shuffle(data)

features=[]

labels=[]

for feature ,label in data:
    features.append(feature)
    labels.append(label)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, labels, test_size=0.20)
```

```python

plt.plot(xtrain,ytrain)
plt.show()
```

```python
model=SVC(kernel='poly',degree=7)
model.fit(xtrain,ytrain)
```

```python

y_hat = model.predict(xtest)

accu= accuracy_score(ytest,y_hat)

print('Accuracy : ',accu)
```
